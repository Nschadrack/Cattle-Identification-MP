```python
def split_dataset(source_root, destination_dir, class_dirs):
    """
    The function which splits the dataset into train, validation and test
    """
    # Iterate through each class directory and split the data
    for class_dir in class_dirs:
        source_class_data_dir = os.path.join(source_root, class_dir)
        destin_class_data_dir = os.path.join(destination_dir, class_dir)


        os.makedirs(source_class_data_dir, exist_ok=True)
        os.makedirs(destin_class_data_dir, exist_ok=True)

        class_files = os.listdir(source_class_data_dir)


        for file in class_files:
            src = os.path.join(source_class_data_dir, file)
            dst = os.path.join(destin_class_data_dir, file)
            shutil.copy(src, dst)
            print(f"Copied: {src} to {dst}")
```

```python
# List all subdirectories (each subdirectory corresponds to a class)
class_dirs = sorted([d for d in os.listdir(train_dir) if os.path.isdir(os.path.join(train_dir, d))])
train_valid_indices = [25, 50, 100, 180, 210]
test_feature_ex_indices = [210, 220, 230, 250, 268]

# Train folders
train_directories = [train_25_dir, train_50_dir, train_100_dir, train_180_dir, train_210_dir]

# Valid directories
valid_directories = [valid_25_dir, valid_50_dir, valid_100_dir, valid_180_dir, valid_210_dir]


# Tesing directories
test_directories = [test_25_dir, test_50_dir, test_100_dir, test_125_dir, test_150_dir,
                    test_180_dir, test_200_dir, test_210_dir, test_220_dir, test_230_dir, test_250_dir, test_268_dir]

# Feature Extractor directories
feature_extractor_directories = [feature_extractor_210_dir, feature_extractor_220_dir,
                                 feature_extractor_230_dir, feature_extractor_250_dir, feature_extractor_268_dir]
```

```python
# Splitting Training
for index, folder in enumerate(train_directories):
  target_classes = class_dirs[:train_valid_indices[index]]
  split_dataset(train_dir, folder, target_classes)
```

```python
# Splitting Validation
for index, folder in enumerate(valid_directories):
  target_classes = class_dirs[:train_valid_indices[index]]
  split_dataset(val_dir, folder, target_classes)
```

```python
# Splitting Testing
for index, folder in enumerate(test_directories):
  target_classes = class_dirs[:test_feature_ex_indices[index]]
  split_dataset(test_dir, folder, target_classes)
```

```python
# Splitting Feature Extractor
for index, folder in enumerate(feature_extractor_directories):
  target_classes = class_dirs[:test_feature_ex_indices[index]]
  split_dataset(feature_extractor_dir, folder, target_classes)
```
