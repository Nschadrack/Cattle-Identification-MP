```python
experiments = os.listdir(experiments_logs_path)
experiments_dicts = []
for experiment in experiments:
  experiment_path = os.path.join(experiments_logs_path, experiment)
  if "accuracy" in experiment.lower():
    with open(experiment_path, 'r') as f:
      experiment = json.load(f)
      experiments_dicts.append(experiment)
    print(experiment)
```

```python
experiments_dicts
```

```python
data = []
for experiment in experiments_dicts:
  X = []
  y = []
  for key, value in experiment.items():
    X.append(int(key))
    y.append(round(value[1], 2))
  data.append([X, y])
```

# Model trained on 25 classes, and adding new classes after training

```python
sch_100 = (data[2][0], np.array(data[2][1])/100)
sch_180 = (data[3][0], np.array(data[3][1])/100)

data = [sch_100, sch_180]

chucks_100 = (df_100["Number of Classes"], df_100["Test Accuracy"])
chucks_180 = (df_180["Number of Classes"], df_180["Test Accuracy"])

data2 = [chucks_100, chucks_180]
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15, 7), sharey=False)

for i in range(2):
  axes[i].plot(data[i][0], data[i][1], marker="o", label="Dot product")
  axes[i].plot(data2[i][0], data2[i][1], marker="o", label="SIAMESE")
  axes[i].set_title(f"Model trained on {data[i][0][0]} classes, and adding new classes")
  # axes[i].grid(True)
  axes[i].legend()
  axes[i].set_xlabel("Number of classes")
  axes[i].set_ylabel("Accuracy")
  # axes[i].set_ylim(0, 1)
  plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=5, figsize=(25, 3), sharey=True)

for i in range(5):
    axes[i].plot(data[i][0], np.array(data[i][1])/100, marker="x")
    axes[i].set_title(f"Model trained on {data[i][0][0]} classes, and adding new classes")
    axes[i].grid(True)
    axes[i].set_xlabel("Number of classes")
    axes[i].set_ylabel("Accuracy")
    # axes[i].set_ylim(0, 1)
plt.tight_layout()
plt.show()
```
